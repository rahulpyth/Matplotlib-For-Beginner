```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ML_Student_Age = np.random.randint(18,45, (100))
Py_Student_Age = np.random.randint(14,40, (100))
```

```python
print(ML_Student_Age)
print(Py_Student_Age)
```

```python
plt.hist(ML_Student_Age)

plt.title("ML Student Age Histogram")
plt.xlabel("Student Age Catogary")
plt.ylabel("Number of Student")
plt.show
```

```python
bins = [15,20,25,30,35,40,45] #bins
plt.hist(ML_Student_Age, bins, rwidth = [0.8], histtype = "bar", orientation = "horizontal", color = "g", label = "ML Student") # 0.8 means 80%

plt.title("ML Student Age Histogram")
plt.xlabel("Student Age Catogary")
plt.ylabel("Number of Student")
plt.legend(loc = 4)
plt.show
```

```python
bins = [15,20,25,30,35,40,45] #bins

plt.figure(figsize = (15,7))

plt.hist([ML_Student_Age, Py_Student_Age], bins, rwidth = 0.8, histtype = "bar", orientation = "vertical", color = ["g", "r"], label = ["ML Student", "Py Student"]) 
#plt.hist(Py_Student_Age, bins, rwidth = [0.8], histtype = "bar", orientation = "vertical", color = "m", label = "Py Student")# 0.8 means 80%

plt.title("ML Student Age Histogram")
plt.xlabel("Student Age Catogary")
plt.ylabel("Number of Student")
plt.legend(loc = 2)
plt.show
```

```python
from matplotlib import style
style.use("ggplot")

plt.figure(figsize = (15,7))

plt.hist([ML_Student_Age, Py_Student_Age], bins, rwidth = 0.8, histtype = "bar", orientation = "vertical", color = ["g", "r"], label = ["ML Student", "Py Student"]) 
#plt.hist(Py_Student_Age, bins, rwidth = [0.8], histtype = "bar", orientation = "vertical", color = "m", label = "Py Student")# 0.8 means 80%

plt.title("ML Student Age Histogram")
plt.xlabel("Student Age Catogary")
plt.ylabel("Number of Student")
plt.legend(loc = 2)
plt.grid(color = "k", linestyle = ":", linewidth = "2")
plt.show
```
