```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
Classes = ["Python", "R", "AI", "ML", "DS"]
class1_Student = [30, 10, 20, 25, 10]
class2_Student = [40, 5, 20, 20, 10]
class3_Student = [35, 5, 30, 15, 15]
```

```python
plt.bar(Classes, class1_Student)
```

```python
plt.barh(Classes, class1_Student) # barh for horizontal
```

```python
style.use("ggplot")
plt.figure(figsize = (10,5))

plt.bar(Classes, class1_Student, width = 0.4, align = "edge", color = "m", edgecolor = "k", linewidth = 2, linestyle = ":", label = "Classes of Student")

plt.legend(loc = 1)
plt.grid(color = "k")
```

```python
plt.figure(figsize = (10,7))

plt.barh(Classes, class1_Student, align = "edge", color = "m", edgecolor = "k", linewidth = 2, linestyle = ":", label = "Classes of Student")

plt.legend(loc = 1)
plt.grid(color = "k")
```

```python
plt.figure(figsize = (10,5))

plt.bar(Classes, class1_Student, width = 0.4, align = "edge", color = "m", linestyle = ":", edgecolor = "k", linewidth = 2, label = "Classes of Student")

plt.title("Bar Chart of 10th Class", fontsize = 20)
plt.xlabel("Subject", fontsize = 20)
plt.ylabel("Number of Student", fontsize = 20)
plt.legend(loc = 1)
plt.grid(color = "k")
plt.show()
```

```python
plt.figure(figsize = (10,5))

plt.bar(Classes, class1_Student, width = 0.2, align = "edge", color = "g", linestyle = ":", linewidth = 2, label = "Classes 1 Student")

plt.bar(Classes, class2_Student, width = 0.2, align = "edge", color = "m", linestyle = ":", linewidth = 2, label = "Classes 2 Student")

plt.title("Bar Chart of 10th Class", fontsize = 20)
plt.xlabel("Subject", fontsize = 20)
plt.ylabel("Number of Student", fontsize = 20)
plt.legend(loc = 1)
plt.grid(color = "k")
plt.show()
```

```python
plt.figure(figsize = (10,5))

Classes_index = np.arange(len(Classes))

width = 0.2

plt.bar(Classes_index, class1_Student, width, align = "edge", color = "g", linestyle = ":", linewidth = 1, label = "Classes 1 Student")

plt.bar(Classes_index + width, class2_Student, width, align = "edge", color = "m", linestyle = ":", linewidth = 1, label = "Classes 2 Student")

plt.xticks(Classes_index + width, Classes, rotation = 30)
plt.title("Bar Chart of 10th Class", fontsize = 20)
plt.xlabel("Subject", fontsize = 20)
plt.ylabel("Number of Student", fontsize = 20)
plt.legend(loc = 1)
plt.grid(color = "k")
plt.show()
```

```python
plt.figure(figsize = (10,5))

Classes_index = np.arange(len(Classes))

width = 0.2

plt.bar(Classes_index, class1_Student, width, align = "edge", color = "g", linestyle = ":", linewidth = 2, label = "Classes 1 Student")

plt.bar(Classes_index + width, class2_Student, width, align = "edge", color = "m", linestyle = ":", linewidth = 2, label = "Classes 2 Student")

plt.bar(Classes_index + width + width, class3_Student, width, align = "edge", color = "b", linestyle = ":", linewidth = 2, label = "Classes 3 Student")

plt.xticks(Classes_index + width, Classes, rotation = 30)
plt.title("Bar Chart of 10th Class", fontsize = 20)
plt.xlabel("Subject", fontsize = 20)
plt.ylabel("Number of Student", fontsize = 20)
plt.legend(loc = 1)
plt.grid(color = "k")
plt.show()
```

```python
plt.figure(figsize = (10,5))

Classes_index = np.arange(len(Classes))

width = 0.2

plt.barh(Classes_index, class1_Student, width, align = "edge", color = "g", linestyle = ":", linewidth = 2, label = "Classes 1 Student")

plt.barh(Classes_index + width, class2_Student, width, align = "edge", color = "m", linestyle = ":", linewidth = 2, label = "Classes 2 Student")

plt.barh(Classes_index + width + width, class3_Student, width, align = "edge", color = "b", linestyle = ":", linewidth = 2, label = "Classes 3 Student")

plt.yticks(Classes_index + width, Classes, rotation = 30)
plt.title("Bar Chart of 10th Class", fontsize = 20)
plt.ylabel("Subject", fontsize = 20)
plt.xlabel("Number of Student", fontsize = 20)
plt.legend(loc = 1)
plt.grid(color = "k")
plt.show()
```
