```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
Classes = ["Python", "R", "AI", "ML", "DS"]
Class1_Student = [30, 10, 20, 25, 10]
Class2_Student = [40, 5, 20, 20, 10]
Class3_Student = [35, 5, 30, 15, 15]
```

```python
style.use("ggplot")
plt.figure(figsize = (10,5))

Classes_index = np.arange(len(Classes))

width = 0.2

plt.bar(Classes_index, class1_Student, width, color = "g", linewidth = 2, align = "edge", label = "Class 1 Student")

plt.bar(Classes_index + width, class2_Student, width, color = "m", linewidth = 2, align = "edge", label = "Class 2 Student")

plt.bar(Classes_index + width + width, class3_Student, width, color = "b", linewidth = 2, align = "edge", label = "Class 3 Student")

plt.xticks(Classes_index + width, Classes, rotation = 20)
plt.title("Bar Chart of 10th Class")
plt.xlabel("Subject", fontsize = 20)
plt.ylabel("Number of Student", fontsize = 20)
plt.legend(loc = 1)
plt.grid(color = "k")

plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize = (10,5))

Classes_index = np.arange(len(Classes))

width = 0.2

plt.barh(Classes_index, class1_Student, width, color = "g", linewidth = 2, align = "edge", label = "Class 1 Student")

plt.barh(Classes_index + width, class2_Student, width, color = "m", linewidth = 2, align = "edge", label = "Class 2 Student")

plt.barh(Classes_index + width + width, class3_Student, width, color = "b", linewidth = 2, align = "edge", label = "Class 3 Student")

plt.yticks(Classes_index + width, Classes, rotation = 20)
plt.title("Bar Chart of 10th Class")
plt.ylabel("Subject", fontsize = 20)
plt.xlabel("Number of Student", fontsize = 20)
plt.legend(loc = 1)
plt.grid(color = "k")

plt.show()
```
