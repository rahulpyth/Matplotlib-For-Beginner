```python
x = df_google_play_store["Rating"]
y = df_google_play_store["Reviews"]
```

```python
from matplotlib import style

style.use("ggplot")

plt.figure(figsize = (15,8))

plt.scatter(x, y, color = "b", marker = "o", s = 100, alpha = 0.5,)

plt.title("Google Play Store Apps", fontsize = 25)
plt.xlabel("Rating", fontsize = 25)
plt.ylabel("Reviews", fontsize = 25)
```
