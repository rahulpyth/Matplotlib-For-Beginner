```python
import matplotlib.pyplot as plt
```

```python
Classes = ["Python", "R", "ML", "AI", "DC"]
Class1_Student = [45, 15, 35, 25, 30]

plt.pie([1])

plt.show()
```

```python
Classes = ["Python", "R", "ML", "AI", "DC"]
Class1_Student = [45, 15, 35, 25, 30]

plt.pie(Class1_Student, labels = Classes)

plt.title("Class 1 Students")
plt.show()
```

```python
Classes = ["Python", "R", "ML", "AI", "DC"]
Class1_Student = [45, 15, 35, 25, 30]

explode = [0.1, 0, 0.2, 0, 0]
colors = ["c", "b", "r", "m","g"]
textprops = {"fontsize":15}


plt.pie(Class1_Student, labels = Classes, explode = explode, colors = colors, autopct = "%0.1f%%", shadow = True, radius = 1.5, startangle = 180,
       textprops = textprops,)
plt.legend(loc = 1)
plt.show()
```
